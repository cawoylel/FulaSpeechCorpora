```python
%%shell

batch_dialects_1="adamawa borgu"
batch_dialects_2="bororro caka"
batch_dialects_3="liptako maacina"
batch_dialects_4="caka"

for dialect in bible/raw/*; do
  dialect_name="$(basename "$dialect")"
  if [[ $batch_dialects_4 =~ (^|[[:space:]])"$dialect_name"($|[[:space:]]) ]] ; then
    output_path=bible/aligned/$dialect_name/
    rm -rf $output_path
    mkdir -p $output_path
    for text in $dialect/*.txt; do
      filename="$(basename "$text")"
      stem=${filename%.*}
      text_output_path=$output_path/$stem
      python fairseq/examples/mms/data_prep/align_and_segment.py --audio_filepath $dialect/$stem.wav --text_filepath $dialect/$stem.txt --lang ful --outdir $text_output_path --uroman uroman/bin
    done
  fi
done
```
